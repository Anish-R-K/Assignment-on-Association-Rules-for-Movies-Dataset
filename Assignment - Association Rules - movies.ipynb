```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movies=pd.read_csv('my_movies.csv',sep=',')
movies
```

```python
movies=movies.replace(np.nan,'',regex=True)
movies
```

```python
movies=movies.iloc[:,5:]
movies
```

```python
# Finding out the frequently watched movies
from mlxtend.frequent_patterns import apriori,association_rules
freq_items=apriori(movies,min_support=0.005,max_len=4,use_colnames=True)
freq_items
```

```python
freq_items.sort_values('support',ascending=False,inplace=True)
freq_items
```

```python
# Pairing the frequently watched movies 
rules=association_rules(freq_items,metric='lift',min_threshold=1)
rules
```

```python
rules.sort_values('lift',ascending=False)
```
